## 2. Cleaning

```python
df = df_raw.copy()

# 1. Drop unusable column
df.drop(columns=['neighbourhood_group'], inplace=True)

# 2. Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# 3. Normalize text
df['neighbourhood'] = df['neighbourhood'].str.title().str.strip()
df['license'] = df['license'].str.strip()

# 4. Drop rows without price (not usable for price analysis)
df = df.dropna(subset=['price'])
print(f'Rows after dropping price NaN: {len(df)}')

# 5. Remove price outliers (> 99th percentile = EUR1,200)
p99 = df['price'].quantile(0.99)
df = df[df['price'] <= p99].copy()
print(f'Rows after removing price outliers (> EUR{p99:.0f}): {len(df)}')

# 6. Cap minimum_nights at 365
df['minimum_nights'] = df['minimum_nights'].clip(upper=365)

# 7. Derived columns useful for analysis
df['is_entire_home']      = df['room_type'] == 'Entire home/apt'
df['is_professional_host']= df['calculated_host_listings_count'] > 5
df['is_long_term']        = df['minimum_nights'] >= 30
df['is_active']           = df['number_of_reviews_ltm'] > 0
df['has_license']         = df['license'].notna()

print(f'\nClean dataset: {df.shape}')
df.head(3)
```

```python
# Data quality summary after cleaning
summary = {
    'Total listings (clean)': len(df),
    '% entire homes': f"{df['is_entire_home'].mean()*100:.1f}%",
    '% professional hosts (>5 listings)': f"{df['is_professional_host'].mean()*100:.1f}%",
    '% long-term (min_nights >= 30)': f"{df['is_long_term'].mean()*100:.1f}%",
    '% active (at least 1 review in last year)': f"{df['is_active'].mean()*100:.1f}%",
    '% with license': f"{df['has_license'].mean()*100:.1f}%",
    'Median price (EUR/night)': f"EUR{df['price'].median():.0f}",
    'Average price (EUR/night)': f"EUR{df['price'].mean():.0f}",
}
pd.Series(summary).rename('Value').to_frame()
```

## 5. EDA - Host Structure

```python
def host_tier(n):
    if n == 1:   return '1 (private)'
    elif n <= 5: return '2-5'
    elif n <= 20:return '6-20'
    else:        return '>20 (property manager)'

df['host_tier'] = df['calculated_host_listings_count'].apply(host_tier)
tier_order = ['1 (private)', '2-5', '6-20', '>20 (property manager)']
tier_counts = df['host_tier'].value_counts().reindex(tier_order)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Host tier pie chart
colors_tier = ['#185FA5', '#1D9E75', '#BA7517', '#D85A30']
axes[0].pie(tier_counts, labels=tier_order, autopct='%1.1f%%',
            colors=colors_tier, startangle=90,
            textprops={'fontsize': 9}, pctdistance=0.75)
axes[0].set_title('Listings by host type', fontsize=12)

# Minimum nights breakdown
def nights_tier(n):
    if n <= 6:   return '1-6 nights'
    elif n <= 29:return '7-29 nights'
    else:        return '>=30 nights'

df['nights_tier'] = df['minimum_nights'].apply(nights_tier)
nights_order = ['1-6 nights', '7-29 nights', '>=30 nights']
nights_counts = df['nights_tier'].value_counts().reindex(nights_order)
colors_nights = ['#185FA5', '#1D9E75', '#BA7517']

axes[1].bar(nights_order, nights_counts, color=colors_nights, width=0.5)
for i, v in enumerate(nights_counts):
    axes[1].text(i, v + 50, f'{v/len(df)*100:.1f}%', ha='center', fontsize=9)
axes[1].set_title('Minimum nights distribution', fontsize=12)
axes[1].set_ylabel('Number of listings')

plt.tight_layout()
plt.savefig('../Output/plots/fig3_host_structure.png', bbox_inches='tight')
plt.show()

print(f"Professional hosts (>5 listings): {(df['calculated_host_listings_count']>5).mean()*100:.1f}% of listings")
print(f"Long-term listings (>=30 days): {df['is_long_term'].mean()*100:.1f}%")
```

## 6. EDA - Licenses and Regulatory Compliance

```python
total = len(df)
has_lic = df['has_license'].sum()
no_lic  = total - has_lic

# Categorize licenses
def lic_category(val):
    if pd.isna(val): return 'Missing'
    v = str(val).strip()
    if v.upper().startswith('IT015146'): return 'IT License (valid)'
    if 'exempt' in v.lower(): return 'Exempt'
    if 'applied' in v.lower(): return 'Pending'
    return 'Other/Invalid'

df['lic_category'] = df['license'].apply(lic_category)
lic_counts = df['lic_category'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

colors_lic = {'IT License (valid)': '#1D9E75', 'Missing': '#D85A30',
              'Pending': '#BA7517', 'Exempt': '#185FA5', 'Other/Invalid': '#888780'}
c_list = [colors_lic.get(k, 'gray') for k in lic_counts.index]

axes[0].pie(lic_counts, labels=lic_counts.index, autopct='%1.1f%%',
            colors=c_list, startangle=90, textprops={'fontsize': 9}, pctdistance=0.8)
axes[0].set_title('License status (Property identification code)', fontsize=11)

# Median price by license status
price_by_lic = df.groupby('lic_category')['price'].median().sort_values(ascending=False)
axes[1].barh(price_by_lic.index, price_by_lic.values,
             color=[colors_lic.get(k, 'gray') for k in price_by_lic.index])
axes[1].set_title('Median price by license status', fontsize=11)
axes[1].set_xlabel('EUR/night')
for i, v in enumerate(price_by_lic.values):
    axes[1].text(v + 1, i, f'EUR{v:.0f}', va='center', fontsize=9)

plt.tight_layout()
plt.savefig('../Output/plots/fig4_license_analysis.png', bbox_inches='tight')
plt.show()

print(lic_counts)
```

## 10. Cleaning Decisions Summary

| Operation | Rows/columns involved | Rationale |
|---|---|---|
| Drop `neighbourhood_group` | 1 column | 100% NaN, no informative value |
| Drop rows without `price` | 2,395 rows | Not usable for economic analysis |
| Remove price outliers > EUR1,200 (99th pct) | 196 rows | They distort averages and visualizations |
| Cap `minimum_nights` at 365 | values > 365 | Reasonable maximum for short-term rental analysis |
| Normalize `neighbourhood` | case + strip | Consistency for joins and aggregations |
| Normalize `license` | strip | Remove extra spaces |
| Add derived columns | 5 columns | `is_entire_home`, `is_professional_host`, `is_long_term`, `is_active`, `has_license` |

**Final dataset**: 20,357 rows x 22 columns
