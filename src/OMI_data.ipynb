# OMI Rental Data - Merge and EDA
This notebook imports OMI rental datasets from multiple semesters between 2018 and 2024, with the goal of building a single consolidated table and starting an exploratory data analysis that can later be integrated with the Airbnb analysis.

```python
import os
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
plt.rcParams.update({'figure.dpi': 120, 'figure.figsize': (10, 4)})
```

---
## 0) Import libraries and datasets
Load all OMI rental CSV files and inspect what is available.

```python
files = sorted(glob.glob('../Data/quotazioni_omi_locazione_*.csv'))
print(f'Files found: {len(files)}')
for f in files:
    print('-', os.path.basename(f))

omi_data = {}
for f in files:
    year_semester = os.path.basename(f).replace('quotazioni_omi_locazione_', '').replace('.csv', '')
    # Most files use header=0, but some releases include a banner row and need header=1
    df_tmp = pd.read_csv(f, sep=';', encoding='latin1', decimal=',', header=0)
    first_col = str(df_tmp.columns[0]) if len(df_tmp.columns) else ''
    unnamed_count = sum(str(c).startswith('Unnamed:') for c in df_tmp.columns)
    if 'Quotazioni Immobiliari' in first_col or unnamed_count > 10:
        df_tmp = pd.read_csv(f, sep=';', encoding='latin1', decimal=',', header=1)
    df_tmp.columns = [c.strip() if isinstance(c, str) else c for c in df_tmp.columns]
    omi_data[year_semester] = df_tmp

# Use the first clean schema as canonical reference
canonical_key = next(k for k, v in omi_data.items() if not any(str(c).startswith('Unnamed:') for c in v.columns))
canonical_columns = list(omi_data[canonical_key].columns)
print('Canonical schema from:', canonical_key)
```

---
## 1) Merging the datasets
Verify the structure and then concatenate all semester files into one consistent dataset.

```python
# Some yearly files expose different headers; use positional alignment for a robust merge
col_counts = {k: df_k.shape[1] for k, df_k in omi_data.items()}
min_cols = min(col_counts.values())
max_cols = max(col_counts.values())

print('Column count range across files:', min_cols, 'to', max_cols)
print('Column counts by file:')
print(col_counts)
```

### Merge execution
Append all semester files, add `Year` and `Semester`, and save one consolidated CSV.

```python
dfs_to_merge = []
for key, df_k in omi_data.items():
    year, semester = key.split('_')
    # Align all files to the canonical schema and preserve consistent column names
    df_tagged = df_k.iloc[:, :len(canonical_columns)].copy()
    df_tagged.columns = canonical_columns
    df_tagged['Year'] = int(year)
    df_tagged['Semester'] = int(semester)
    dfs_to_merge.append(df_tagged)

df_total = pd.concat(dfs_to_merge, ignore_index=True)
print('Merged rows:', len(df_total))
print('Merged columns:', len(df_total.columns))
display(df_total.head(5))

output_path = '../Data/total_rentals.csv'
df_total.to_csv(output_path, index=False, sep=';', encoding='latin1')
print(f'Saved merged dataset to: {output_path}')
```

---
## 2) Preprocessing
After merging, check missing values and remove columns with a single unique value because they are not informative.

```python
print('Merged dataset shape before preprocessing:', df_total.shape)

missing = df_total.isna().sum()
total_missing = missing.sum()
print(f'Total missing values: {total_missing} ({(total_missing / df_total.size * 100):.2f}%)')

single_value_cols = [c for c in df_total.columns if c not in ['Year', 'Semester'] and df_total[c].nunique(dropna=False) <= 1]
print('Single-unique columns to drop:', single_value_cols)

if single_value_cols:
    df_total = df_total.drop(columns=single_value_cols).copy()

print('Shape after preprocessing:', df_total.shape)
```

---
## 3) EDA - start
Begin exploratory analysis on the merged and preprocessed dataset, following the same approach used in the Airbnb notebook.

```python
for col in df_total.columns:
    unique_vals = df_total[col].dropna().unique()
    print(f'Column: {col} has {len(unique_vals)} unique values and has type {df_total[col].dtype}')
```

```python
print('Descriptive statistics for numeric columns:')
display(df_total.describe(include=[np.number]).T)
```

### 3.1 Zone and Fascia overview
Zone and Fascia are the key geographic and market dimensions for the later integration step, so we inspect them before plotting trends.

```python
zone_counts = df_total['Zona'].value_counts()
fascia_counts = df_total['Fascia'].value_counts()

print('Number of unique zones:', df_total['Zona'].nunique())
print('Number of unique fascia values:', df_total['Fascia'].nunique())

display(zone_counts.head(10).to_frame('count'))
display(fascia_counts.to_frame('count'))

zone_fascia_summary = (
    df_total.groupby(['Zona', 'Fascia'])
    .agg(
        n_records=('Year', 'size'),
        avg_loc_min=('Loc_min', 'mean'),
        avg_loc_max=('Loc_max', 'mean')
    )
    .reset_index()
    .sort_values('n_records', ascending=False)
 )
display(zone_fascia_summary.head(15))

# Save the zones unique values to a text file
zone_values = df_total['Zona'].dropna().unique()
with open('../Output/unique_zones.txt', 'w') as f:
    for zone in zone_values:
        f.write(f"{zone}\n")
```

### 3.2 Average min and max loc by period
The next plot tracks the average `Loc_min` and `Loc_max` across years and semesters.

```python
period_summary = (
    df_total.groupby(['Year', 'Semester'])
    .agg(avg_loc_min=('Loc_min', 'mean'), avg_loc_max=('Loc_max', 'mean'))
    .reset_index()
    .sort_values(['Year', 'Semester'])
 )
period_summary['Period'] = period_summary['Year'].astype(str) + ' S' + period_summary['Semester'].astype(str)

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(period_summary['Period'], period_summary['avg_loc_min'], marker='o', linewidth=2, label='Average min loc')
ax.plot(period_summary['Period'], period_summary['avg_loc_max'], marker='o', linewidth=2, label='Average max loc')
ax.set_title('Average min and max loc by year and semester')
ax.set_xlabel('Year / semester')
ax.set_ylabel('Average loc value')
ax.tick_params(axis='x', rotation=45)
ax.legend()
plt.tight_layout()
plt.show()
```

### 3.3 Evolution by fascia
This plot compares how `Loc_min` and `Loc_max` evolve over time for each `Fascia`.

```python
fascia_summary = (
    df_total.groupby(['Year', 'Semester', 'Fascia'])
    .agg(avg_loc_min=('Loc_min', 'mean'), avg_loc_max=('Loc_max', 'mean'))
    .reset_index()
    .sort_values(['Year', 'Semester'])
 )
fascia_summary['Period'] = fascia_summary['Year'].astype(str) + ' S' + fascia_summary['Semester'].astype(str)

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharex=True)

sns.lineplot(data=fascia_summary, x='Period', y='avg_loc_min', hue='Fascia', marker='o', ax=axes[0])
axes[0].set_title('Evolution of average min loc by fascia')
axes[0].set_xlabel('Year / semester')
axes[0].set_ylabel('Average min loc')
axes[0].tick_params(axis='x', rotation=45)

sns.lineplot(data=fascia_summary, x='Period', y='avg_loc_max', hue='Fascia', marker='o', ax=axes[1])
axes[1].set_title('Evolution of average max loc by fascia')
axes[1].set_xlabel('Year / semester')
axes[1].set_ylabel('Average max loc')
axes[1].tick_params(axis='x', rotation=45)

for ax in axes:
    ax.legend(title='Fascia')

plt.tight_layout()
plt.show()
```
