```python
import pandas as pd

# 1. Gather all dataframes, add Year/Semester, and put them in a list
dfs_to_merge = [
    globals()[name].assign(Year=name.split('_')[1], Semester=name.split('_')[2])
    for name in list(globals().keys()) if name.startswith('df_20') and len(name) == 9
]

df_total = pd.concat(dfs_to_merge, ignore_index=True)

print(len(df_total))
display(df_total.head(10))
```

```python
# Save the merged dataframe to the Data folder
df_total.to_csv('Data/total_rentals.csv', index=False, sep=';', encoding='latin1')

print("File 'total_rentals.csv' saved successfully in the Data folder!")
```
